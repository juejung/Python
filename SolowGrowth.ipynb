```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This temporary script file is located here:
C:\Users\jjung\.spyder2\.temp.py
"""

import time                     # Imports system time module to time your script
from pylab import *             # Imports numpy, scipy, and matplotlib etc.
from scipy.optimize import fsolve
import scipy.optimize as optimize
%matplotlib inline  

tic = time.clock()              # Start timer

close('all')
# -----------------------------------------------------------------------------

# -----------------------------------------------------------------------------
# Parameters
# -----------------------------------------------------------------------------

n = 0.02
d = 0.10
z = 1.0
alpha = 0.3
s = 0.15

# -----------------------------------------------------------------------------
# Function definitions
# -----------------------------------------------------------------------------
# 0
def f_y(k):
    y = k**alpha
    return y
# 1
def f_kstar(k):
    return s*z*f_y(k)/(1+n) + (1-d)/(1+n)*k

# 2.1
def f_left(k):
    return s*z*f_y(k)
    
# 2.2
def f_right(k):
    return (n+d)*k
    
# 3
def f_cstar(k):
    return z*f_y(k)-(n+d)*k

def f_cstarNeg(k):
    return -(z*f_y(k)-(n+d)*k)
    
# Root function
def f_zero(k):
    return f_left(k)-f_right(k)

# -----------------------------------------------------------------------------
# Root finding: find steady state level of capital kstar
# -----------------------------------------------------------------------------
print " -------------- Fsolve ------------"
kstar = fsolve(f_zero, 1.5)[0] # starting from x = 2
print "Steady state k*= ", kstar

print " --------- Fmin the negative consumption function"
kmax = optimize.fmin(f_cstarNeg, 3.0)[0] # starting from x = 2
print "kmax that maximizes consumption c*= ", kmax
    
# -----------------------------------------------------------------------------
# Function plots
# -----------------------------------------------------------------------------
kv = arange(0,3,0.01)
    
    
fig1 = figure()
subplot(221)   
plot(kv,kv,'--',kv,f_kstar(kv))  
axvline(x=kstar, ymin=0, ymax=f_kstar(kstar),  color='r')
title('Law of motion of capital')
xlabel('k')
ylabel("k'")
legend(('$45^o$', "k'"),'lower right', shadow=True)
  
subplot(222) # creates 2nd subplot with yellow background
plot(kv,f_left(kv),'r--', kv,z*f_y(kv),'k-', kv,f_right(kv),'b-')
axvline(x=kstar, ymin=0, ymax=0.4,  color='r')
title('Steady state: k*')
xlabel('k')
legend(('szf(k*)','z*f_y(k)','(n+d)k*'),'upper left', shadow=True)
   
subplot(223) # creates 2nd subplot with yellow background
plot(kv,f_left(kv),'r--',kv,f_right(kv),'b-')
axvline(x=kstar, ymin=0, ymax=0.4,  color='r')
title('Steady state: k*')
xlabel('k')
legend(('szf(k*)', '(n+d)k*'),'lower right', shadow=True)
  
kv = arange(0,8,0.1)  
subplot(224) # creates 2nd subplot with yellow background
plot(kv,f_cstar(kv) ,'b-')
axvline(x=kmax, ymin=0, ymax=1.0,  color='r')
title('Consumption: c*')  
xlabel('k')
ylabel('c')
show()

 
```
